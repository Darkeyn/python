```python
from array import *
from faker import Faker
fake = Faker()
temerature = [0] * 50
years = []
for i in range(50):
    temerature[i] = [0] * 0
    years.extend([i + 1972])
for i in range(50):
    sum = 0
    for j in range(12):
        temerature[i].extend([fake.pyfloat(right_digits=1, min_value=-40, max_value=40)])
print(temerature)
print(len(temerature))
print(years)
```

```python
average = []
EMAaverage = []
for i in range(50):
    sum = 0
    for j in range(12):
        sum += temerature[i][j]
    average.extend([round(sum / 12, 1)])
    if i < 1:
        EMAaverage.extend([round(sum / 12, 1)])
    elif i > 0:
        EMAaverage.extend([round(((sum / 12)*(2 / (1 + i)) + (EMAaverage[i-1]*(1 - (2 / (1 + i))))), 1)])
    
print(average)
print(EMAaverage)
```

```python
import matplotlib.pyplot as plt
plt.plot(years, average, '.-b', lw=1)
plt.plot(years, EMAaverage, lw=3, color='blue')
plt.grid(True)
plt.show()
```
